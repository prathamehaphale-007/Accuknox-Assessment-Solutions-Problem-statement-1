```python
import requests
import sqlite3
```

```python
url = "https://openlibrary.org/search.json?q=python"
data = requests.get(url).json()
```

```python
books = []
for b in data["docs"][:10]:
    books.append({
        "title": b.get("title"),
        "author": b.get("author_name", ["Unknown"])[0],
        "year": b.get("first_publish_year", 0)
    })
```

```python
conn = sqlite3.connect("books.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY,
    title TEXT,
    author TEXT,
    year INTEGER
)
""")
```

```python
for book in books:
    cursor.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)",
                   (book["title"], book["author"], book["year"]))
```

```python
conn.commit()

cursor.execute("SELECT * FROM books")
print(cursor.fetchall())
```

```python
conn.close()
```
