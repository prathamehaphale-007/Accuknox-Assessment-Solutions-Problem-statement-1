```python
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://69649217e8ce952ce1f21afd.mockapi.io/students"
data = requests.get(url).json()
```

```python
names = []
scores = []
for s in data:
    names.append(s["name"])
    scores.append(float(s["scores"]))
```

```python
total = 0
for s in scores:
    total += s
avg = total / len(scores)
print("Average score:", avg)
```

```python
from numpy import angle
plt.bar(names, scores, color="orange")
plt.axhline(avg, linestyle="dashed")
plt.title("Student Scores")
plt.ylabel("Score")
plt.xlabel("Students")
step = len(names) // 12
plt.xticks(np.arange(0, len(names), step), rotation=45)
plt.show()
```
