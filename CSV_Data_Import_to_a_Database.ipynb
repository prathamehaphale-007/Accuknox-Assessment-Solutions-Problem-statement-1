```python
import sqlite3
import csv
```

```python
conn = sqlite3.connect("users.db")
cur = conn.cursor()

cur.execute("""
CREATE TABLE users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT
)
""")

conn.commit()
```

```python
with open("/content/users.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        cur.execute(
            "INSERT INTO users (name, email) VALUES (?, ?)",
            (row[0], row[1])
        )

conn.commit()
```

```python
cur.execute("SELECT * FROM users")
print(cur.fetchall())
conn.close()
```
